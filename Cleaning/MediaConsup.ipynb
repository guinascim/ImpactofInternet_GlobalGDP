```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/guinascim/ImpactofInternet_GlobalGDP/main/Dataset/Dataset%20Inicial/Media/social_media.csv'
media = pd.read_csv(url)
```

```python
media.head()
```

```python
media['Date'] = pd.to_datetime(media['Date'])
media['Year'] = media['Date'].dt.year
```

```python
media.drop(columns=['Date'], inplace=True)
media = media[['Year'] + [col for col in media if col != 'Year']]
```

```python
media.head()
```

```python
media_grouped = media.groupby('Year').sum().reset_index()
```

```python
media_grouped = media_grouped[(media_grouped['Year'] >= 2010) & (media_grouped['Year'] <= 2020)]
media_grouped.head()
```

```python
media_grouped.shape
```

```python
transposed_df = media_grouped.T.reset_index()
transposed_df.columns = transposed_df.iloc[0]
transposed_df = transposed_df.drop(0)
transposed_df.reset_index(drop=True, inplace=True)
transposed_df = transposed_df.rename(columns={'Year': 'Media'})
transposed_df.head()
```

```python
transposed_df.columns = transposed_df.columns.astype(str).str.split('.').str[0]
```

```python
transposed_df.to_csv('me_popular.csv', index=False)
```
