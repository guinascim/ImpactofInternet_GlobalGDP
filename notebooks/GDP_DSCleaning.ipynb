```python
import pandas as pd
```

```python
url = "https://github.com/guinascim/ImpactofInternet_GlobalGDP/raw/main/Dataset%20Inicial/Media/Globalgdp.csv"
```

```python
gdp = pd.read_csv(url)
```

```python
gdp.head()
```

```python
gdp.shape
```

```python
country_name = gdp["Country Name"]
```

```python
non_numeric_columns = gdp.columns[~gdp.columns.str.isnumeric()]
gdp = gdp.drop(columns=non_numeric_columns)
```

```python
gdp.columns = gdp.columns.astype(int)
gdp = gdp.loc[:, (gdp.columns >= 2010) & (gdp.columns <= 2020)]
```

```python
gdp = pd.concat([country_name, gdp], axis=1)
```

```python
gdp = gdp.rename(columns={'Country Name': 'Country'})
```

```python
for column in gdp.columns[1:]:
    gdp[column] = gdp[column].apply(lambda x: f"{x:.2f}")
```

```python
gdp.to_csv('cleaned_gdp.csv', index=False)
```
