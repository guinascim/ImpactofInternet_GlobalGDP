```python
import pandas as pd
import requests
from io import StringIO

url = 'https://raw.githubusercontent.com/guinascim/ImpactofInternet_GlobalGDP/main/Dataset/Dataset%20Inicial/Internet%20users/share-of-individuals-using-the-internet.csv'

response = requests.get(url)

if response.status_code == 200:
    usage = pd.read_csv(StringIO(response.text))
    print(usage.head())

```

```python
usage.head()
```

```python
usage.columns
```

```python
usage.drop(columns='Code', inplace=True)
usage.rename(columns={'Individuals using the Internet (% of population)': 'Users'}, inplace=True)
usage.rename(columns={'Entity': 'Country'}, inplace=True)
```

```python
usage.head()
```

```python
pivot_df = usage.pivot_table(index='Country', columns='Year', values='Users', aggfunc='sum')
pivot_df = pivot_df.loc[:, 2010:2020]
pivot_df.reset_index(inplace=True)
pivot_df.columns.name = None
pivot_df.head()
```
